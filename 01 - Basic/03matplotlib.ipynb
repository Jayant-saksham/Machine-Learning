```python
# pip install matplotlib
import matplotlib.pyplot as plt
```

**Plot in Matplotlib**

```python
x = [1,2,3,4]
y = [4,3,2,1]
plt.plot(x, y, marker = 'o')
plt.show()
```

```python
import numpy as np
l = [i for i in range(1,50)]
m = [i*i for i in range(1,50)]
l1 = np.array(l)
l2 = np.array(m)
plt.plot(l1, l2)
plt.title("Parabola")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(["y = x*x"])
plt.show()
```

**Histogram**

```python
ml_students = np.random.randint(18,45,100)
py_students = np.random.randint(15,22,100)
```

```python
ml_students
```

```python
py_students
```

```python
from matplotlib import style
style.use("ggplot")
bins = [15,20,25,30,35,40,45]
plt.hist([ml_students, py_students], bins = bins, orientation = "vertical")
plt.title("ML students age histogram")
plt.xlabel("Student age")
plt.ylabel("Number of students")
plt.legend(["ML Students", "Python Students"])
plt.show()
```

**Bar Plot**

```python
sub = ["Python", "R", "AI", "ML", "DSA"]
students = [30,10,50,20,43]
plt.bar(sub, students)
plt.show()
```

**Pie Chart**

```python
sub = ["Python", "R", "AI", "ML", "DSA"]
students = [30,10,50,20,43]
explode = [0,0,0.1,0,0]
plt.pie(students, labels = sub, explode=explode)
plt.show()
```
