# Loss function

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
# We are considering loss function as 
# loss function = Sum((y - y')^2)/n
def gradient_descent(x, y):
    m_current = 0
    c_current = 0
    n = len(x)
    learning_rate = 0.0001
    iterations = 2000
    lossfx = []
    m = []
    c = []
    for i in range(iterations):
        y_predicted = m_current*x + c_current
        loss = (1/n) * sum([i**2 for i in (y-y_predicted)])
        lossfx.append(loss)
        md = -(2/n)*sum(x*(y - y_predicted))
        cd = -(2/n)*sum(y-y_predicted)
        m_current = m_current - learning_rate*md
        c_current = c_current - learning_rate*cd 
        m.append(m_current)
        c.append(c_current)
        print(f"Iteation {i} |  M = {m_current} and C = {c_current} Loss = {loss}")
```

```python
gradient_descent(x=x, y=y)
```
