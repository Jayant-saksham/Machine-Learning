# Learning Rate

**How quickly we are moving is our Learning rate towards the answer**

# Loss function
<img src="https://www.section.io/engineering-education/understanding-loss-functions-in-machine-learning/mean-squared-error.PNG">

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
# We are considering loss function as 
# loss function = Sum((y - y')^2)/n
def gradient_descent(x, y):
    m_current = 0
    c_current = 0
    n = len(x)
    learning_rate = 0.0001
    iterations = 2000
    for i in range(iterations):
        y_predicted = m_current*x + c_current
        md = -(2/n)*sum(x*(y - y_predicted))
        cd = -(2/n)*sum(y-y_predicted)
        m_current = m_current - learning_rate*md
        c_current = c_current - learning_rate*cd 
        print(f"Iteation {i} |  M = {m_current} and C = {c_current}")
```

```python
gradient_descent(x=x, y=y)
```
